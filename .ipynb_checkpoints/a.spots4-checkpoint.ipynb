```r
df1=j1 %>% bind_rows %>% separate_wider_delim('position',',',names=c('x','y')) %>% muat(vars(x,y),asd) %>% 
    mu(V3=createTime %>% maps(~ifelse(sde(.,'-',T), asd(.)/1000, 
        strptime(srm(.,'\\..*'), '%Y-%m-%d %H:%M:%S') %>% asd)) %>% asi) %>% asdf
df12=j2 %>% bind_rows %>% separate_wider_delim('position',',',names=c('x','y')) %>% muat(vars(x,y),asd) %>% 
    mu(V3=createTime %>% maps(~ifelse(sde(.,'-',T), asd(.)/1000, 
        strptime(srm(.,'\\..*'), '%Y-%m-%d %H:%M:%S') %>% asd)) %>% asi) %>% asdf
```

```r
mat=vmap(1:nr(dfs[[1]]) %>% sn,~{
    x=.x
    map(1:nr(dfs[[2]]) %>% sn,~{
        a=dfs[[1]][x,1:2] %>% ul
        b=dfs[[2]][.x,1:2] %>% ul
        sqrt(sum((a-b)^2))
    }) %>% asdf
}) %>% bind_rows
```
